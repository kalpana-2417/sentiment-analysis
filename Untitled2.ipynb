```python
from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
import pandas as pd
```

```python
# Load your travel review dataset
df = pd.read_csv("/content/tarvel review ratings.csv")  # Make sure this file exists
```

```python
# Check column names
print("Columns:", df.columns)

# Example: Use 'User' as text, 'Category 1' as label - REPLACE WITH YOUR ACTUAL COLUMNS
texts = df['User'].astype(str)
labels = df['Category 1']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
import pandas as pd
import numpy as np # Import numpy for digitize

# Assuming 'Category 1' is the column with ratings (0-5)
# Convert continuous ratings into discrete categories (e.g., 5 bins)
bins = np.arange(0, 6, 1) # Create bins: [0, 1, 2, 3, 4, 5]
labels = np.digitize(df['Category 1'], bins) - 1 # Assign each rating to a bin index (0-4)

# Split into training and testing
X_train, X_test, y_train, y_test = train_test_split(texts, labels, test_size=0.2, random_state=42)

# Create pipeline: vectorizer + classifier
model = make_pipeline(CountVectorizer(), MultinomialNB())
model.fit(X_train, y_train)
```

```python
# Interactive prediction
while True:
    user_input = input("Enter a travel review (or 'quit' to exit): ")
    if user_input.lower() == 'quit':
        break
    prediction = model.predict([user_input])[0]
    print(f"Predicted Sentiment: {prediction}")
```
